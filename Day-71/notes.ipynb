```python
clean_data = data.dropna()
clean_data.tail()
```

```python
max_salary = clean_data["Starting Median Salary"].idxmax()
```

```python
clean_data["Undergraduate Major"].loc[max_salary]
```

```python
# Highest Mid-Career Salary
highest_mid = clean_data["Mid-Career 90th Percentile Salary"].idxmax()
print(clean_data["Mid-Career 90th Percentile Salary"].max())
clean_data["Undergraduate Major"][highest_mid]
```

```python
# Lowest Starting Salary
lowest = clean_data["Starting Median Salary"].argmin()
name = clean_data["Undergraduate Major"][lowest]
salary = clean_data["Starting Median Salary"].min()
print(salary)
name
```

```python
# Lowest Mid-Career Salary
lowest_mid = clean_data["Mid-Career Median Salary"].argmin()
name = clean_data["Undergraduate Major"][lowest_mid]
salary = clean_data["Mid-Career Median Salary"][lowest_mid]
print(salary)
name
```

```python
# Low Risk Majors
spread_col = clean_data["Mid-Career 90th Percentile Salary"] - clean_data["Mid-Career 10th Percentile Salary"]
low_risk = clean_data.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
# Top five highest paying potential degrees
sort_df = clean_data.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
sort_df.head()
```
